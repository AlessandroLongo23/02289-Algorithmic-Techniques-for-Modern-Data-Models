## Exercise 2.1.1: PC3 Algorithm
### Problem
#### Part 1
Run the P3C algorithm on the following example:
$$1, 7, 25, 23, 17, 16, 21, 38, 3, 8, 11$$

```python
from lib.p3c import P3C

initial_colors = [1, 7, 25, 23, 17, 16, 21, 38, 3, 8, 11]
p3c = P3C(initial_colors=initial_colors)
final_colors, steps = p3c.run()
print(f"""
final_colors: {[final_colors.key for final_colors in final_colors]}
steps: {steps}
""")

import networkx as nx

G = nx.Graph()
G.add_edges_from([(i, i + 1) for i in range(len(final_colors) - 1)])
for node in G.nodes:
    G.nodes[node]['color'] = final_colors[node].key

colors = {
    0: 'white',
    1: 'blue',
    2: 'red'
}
nx.draw(G, with_labels=True, node_color=[colors[G.nodes[node]['color']] for node in G.nodes], node_size=500, edgecolors='black')
```

#### Part 2
Give an instance where the algorithm PC3 runs in $n$ rounds.

#### Part 2
Any decreasing sequence will run in $n$ rounds, since the only local maxima at round $i$ is at position $i$. Analogous for increasing sequence.

```python
initial_colors = sorted(initial_colors)
p3c = P3C(initial_colors=initial_colors)
final_colors, steps = p3c.run()
print(f"""
final_colors: {[final_colors.key for final_colors in final_colors]}
steps: {steps}
""")
```
