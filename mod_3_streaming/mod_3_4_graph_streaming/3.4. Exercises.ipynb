## Exercise 6: $k$-center clustering
### Problem
In this exercise, we focus on a problem related to graph streaming. We are given a metric space $(X,d)$ and a positive integer $k$. For a subset $Y\subseteq X$ and an $x\in X$, define: 
$$d(Y,x)=d(x,Y)=\min_{y\in Y}d(x,y)$$
if $Y=\emptyset$, define $d(Y,x)=d(x,Y)=\infty$.
Instead of a stream of edges, we get a stream of points from $X:x_1,x_2,\ldots$. Let $S$ denote the set of points from the stream. The output should be a set of centers $Y\subseteq S$, with $|Y|\le k$, such that the cost:
$$\max_i d(x_i,Y)=\max_{x\in S}d(x,Y)$$
is minimized.
Let $\text{OPT}$ denote the cost of an optimal solution. We will analyze a semi-streaming algorithm using $O(k)$ space which achieves a solution of cost at most $r=2\text{OPT}$.
```
proc centers(S):
	Y = {}
	for x in S:
		if d(x,Y) > r:
			Y = Y + x	
```
A big downside of the algorithm is that it needs to know $r$. Fortunately, there is a modification of the algorithm that doesn't have this requirement and which obtains a solution of cost at most $(2+\epsilon)\text{OPT}$ for any chosen constant $\epsilon\gt0$. We will not focus on this.
#### Part 1
Show that the algorithm finds a set of at most $k$ clusters.

#### Part 2
Show that the cost of this solution is at most $r$.
